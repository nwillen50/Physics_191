This notebook is used to produce a graph of critical field vs temperature for a singular metal

Primarily for Type I Superconductors

```python
# Import necessary librarys
import matplotlib.pyplot as plt
import pandas as pd
import glob

from type_I import type_I_analysis
```

```python
# Import mutliple .txt files

# Combine all files for a given sample

# Available samples: (PB2, SN, IN, SNPB, PTFE, IN2BI, IN4BI, CU, PB0, PB1)
# Replace blank in "sc_data/__/*.txt" with sample name

files = glob.glob("meissner/PB2/sc_data/*.txt")


# Create dictionary of superconductor files
sc_data = {}

# read files into dictionary
for file in files:
    sc_data[file] = pd.read_csv(file, sep=r"\s+")
```

```python
k_array, critical_fields = type_I_analysis(sc_data)
```

```python
plt.figure() # Could replace this with plt.subplot(), depending on vibes
plt.scatter(k_array, critical_fields)
plt.xlabel("Temperature (K)")
plt.ylabel("Critical Field (kG)")
plt.title("Critical Field vs Temperature: PB2")


plt.show
```
