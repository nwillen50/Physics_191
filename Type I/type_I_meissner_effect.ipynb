This notebook is used to produce a graph of critical field vs temperature for a singular metal

Primarily for Type I Superconductors

```python
# Import necessary librarys
import pandas as pd
import glob

from type_I import type_I_analysis, meissner_plot
```

```python
# Import mutliple .txt files

# Combine all files for a given sample

# Available samples: (PB2, SN, IN, SNPB, PTFE, IN2BI, IN4BI, CU, PB0, PB1)
# Define metal = sample name

metal = "PB2"

files = glob.glob(f"../meissner/{metal}/sc_data/*.txt")


# Create dictionary of superconductor files
sc_data = {}

# read files into dictionary
for file in files:
    sc_data[file] = pd.read_csv(file, sep=r"\s+")
```

Calculate critical fields and pair them with their associated temperature

```python
k_array, critical_fields = type_I_analysis(sc_data, 0.3, True)
```

Plot data with a best-fit meissner model

```python
# Plot data w/ best fit line
meissner_plot(k_array, critical_fields, metal)
```
