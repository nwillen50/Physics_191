```python
from graphs import env_graph, osc_graph, ft_graph
import matplotlib.pyplot as plt
```

```python
file = "Data/FT/Fluorinated/FC43.csv"
freq1, ft1, _ = ft_graph(file)
file = "Data/FT/Fluorinated/FC70.csv"
freq2, ft2, peaks = ft_graph(file)
file = "Data/FT/Fluorinated/FC770.csv"
freq3, ft3, _ = ft_graph(file)
file = "Data/FT/Fluorinated/HT110.csv"
freq4, ft4, _ = ft_graph(file)
```

```python
fig = plt.figure()
plt.plot(freq1/1000, ft1, color='C0', label="FC-43")
plt.plot(freq2/1000, ft2, color='C1', label="FC-70")
plt.plot(freq3/1000, ft3, color='C2', label="FC-770")
plt.plot(freq4/1000, ft4, color='C3', label="HT-110")
plt.grid()
plt.minorticks_on()
plt.grid(visible=True, which='minor', color='lightgray', linestyle='--', alpha=0.5)
plt.axvline(freq2[peaks[0]]/1000, color='red', linestyle='--', alpha=0.5)
plt.axvline(freq2[peaks[1]]/1000, color='red', linestyle='--', alpha=0.5)
plt.xlim(2.4e1, 2.8e1)
plt.xlabel("Frequency (kHz)")
plt.ylabel("Voltage (V)")
plt.title("Fluorine Compound Resonant Frequencies")


legend1 = plt.legend(loc='upper right', title='Fluorine Compounds')
plt.gca().add_artist(legend1)


plt.legend(handles=[
    plt.Line2D([0], [0], color='red', linestyle='--', label=f'{freq2[peaks[0]]/1000:.2f} kHz'),
    plt.Line2D([0], [0], color='red', linestyle='--', label=f'{freq2[peaks[1]]/1000:.2f} kHz')
], loc='upper left', title='Resonant Frequencies')

fig.savefig(f"fluourine_frequencies.png", dpi=300, bbox_inches="tight")

plt.show()

```
